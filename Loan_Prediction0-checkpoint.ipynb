```python
#in term of creadit_History, we can assume that if they not provide they would never got a credit before
data['Credit_History'].fillna(0,inplace=True)
```

```python
#Same as self_employed, we assume that if they not provide they was unself employed too
data['Self_Employed'].fillna('No',inplace=True)
```

```python
data['Total income']=data['ApplicantIncome']+data['CoapplicantIncome']
```

```python
data['Total income_group']=pd.cut(data['Total income'],3,labels=["Min",'Mid','High'])
data.head()
```

```python
Data_group=data.groupby(['Total income_group']).mean()
Data_group
```

```python
data['LoanAmount'].fillna(data.groupby(['Total income_group'])['LoanAmount'].transform('mean'), inplace = True)
data.head()
```

```python
data['Loan_Amount_Term'].fillna(value = data.Loan_Amount_Term.median(),inplace=True)
data.head()
```
